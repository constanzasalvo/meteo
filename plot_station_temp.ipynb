# Plot Daily Temperature in Qikiqtarjuaq

Example for plotting daily temperature (mean, max and min)

```python
# Libraries
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from meteostat import Point, Daily
from meteostat import Stations
```

```python
#Site of interest
##Qikiqtarjuat:
qikiq_lat = 67.5521451887046
qikiq_lon = -64.02372900795045
qikiq = Point(qikiq_lat, qikiq_lon, 70)

# Set time period
start =  datetime(2023, 9, 1)
end = datetime(2024, 8, 31)

#Weather data
data = Daily(qikiq, start, end)
data = data.fetch()
```

```python
#plot
data.plot(y=['tavg', 'tmin', 'tmax'], figsize=(15, 7))
plt.xlabel('Date')
plt.ylabel('Temperature C')
plt.legend()
plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # Format date display
plt.xlim(f'2023-09-01 00:00:00', '2024-08-31 23:59:00')
plt.xticks(rotation=45)
plt.legend(loc='lower right')
plt.show()
```
